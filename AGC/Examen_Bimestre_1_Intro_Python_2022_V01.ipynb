````
def get_cmap(n, name='hsv'):
  import matplotlib.pyplot as plt
  return plt.cm.get_cmap(name, n)
````

````
def grafCantTipos(datos, desdeAnio, hastaAnio, anotacion=False):
  """
  Desc: Función que se le pasa un DataFrame con anios y tipos o categorías
        y grafica para cada tipo la cantidad según el año de realización
        en los rangos (desdeAnio, hastaAnio) ambos inclusive
  Args:
    datos (df): DataFrame con columna type(str) y columna release_year (int) con formato yyyy
                      Ej:
                                 type   release_year
                          0     Movie           2020
                          6     TV Show         2021
    desdeAnio (int): Año en formato yyyy desde el cual se desea consultar
    hastaAnio (int): Año en formato yyyy hasta el cual se desea consultar inclusive
    anotacion (bool): imprime las anotaciones de los puntos en caso de requerirlo
  Returns:
    N/A 
  """
  import matplotlib.pyplot as plt
  import pandas as pd
  import numpy as np
  import datetime
  from matplotlib.collections import EventCollection
  
  try:
    
    #Validar 
    if (type(desdeAnio) != type(1)):
      return 'desdeAnio: error de tipo, debe ser un int'

    if (type(hastaAnio) != type(1)):
      return 'hastaAnio: error de tipo, debe ser un int'

    if (type(anotacion) != type(True)):
      return 'anotacion: error de tipo, debe ser un bool (True/False)'

    currentDateTime = datetime.datetime.now()
    date = currentDateTime.date()
    year = int(date.strftime("%Y"))

    if (desdeAnio < 1895) or (desdeAnio > year):
      s_error=f'desdeAnio debe estar comprendido entre 1895 y {year}'
      return s_error

    if (hastaAnio < 1895) or (hastaAnio > year+1):
      s_error=f'hastaAnio debe estar comprendido entre 1895 y {year+1}'
      return s_error

    if desdeAnio==hastaAnio:
      s_error='El desdeAnio y hastaAnio no pueden ser iguales'
      return s_error

    if desdeAnio>hastaAnio:
      s_error='El desdeAnio no puede ser mayor a hastaAnio'
      return s_error

    #Procesar    
    datos=pd.DataFrame(datos, columns=['type','release_year']) 

    #Obtener tipos
    s_error='Obtener los tipos'
    s_cat_tipos=datos['type'].drop_duplicates()

    #Filtrar por fechas
    mask2 = (datos['release_year'] >= desdeAnio) & (datos['release_year'] <= hastaAnio)
    datos = datos.loc[mask2]

    #Preparar figure y ax
    s_error='Error al preparar la figura'    
    fig, ax = plt.subplots(1, 1, figsize=(12,5), facecolor='#EBEBEB', edgecolor='black')     
    ax.set(facecolor='lightsteelblue')    
    ax.set_xticks(np.arange(desdeAnio, hastaAnio+1, 1))
    ax.set_xlim(desdeAnio-1,hastaAnio+1)
    
    #Preparar auxiliares
    totales=pd.Series(index=np.arange(desdeAnio, hastaAnio+1, 1), dtype='int64')
    i=0
    cmap = get_cmap(len(s_cat_tipos)*2)

    s_error='Recorrer los tipos y plotearlos'
    for t in s_cat_tipos.values:
      i+=2      
      totales[:]=0
      anios=datos[datos.type.values==t]['release_year'].value_counts().sort_index()
      totales=totales.add(anios, fill_value=0).astype(int)
      
      if anotacion:
        x=(list(totales.index))
        y=(list(totales.values))
        for j in range(len(x)):
          posx=0
          if len(str(y[j]))==1:
            posx=x[j]-0.4
          elif len(str(y[j]))==2:
            posx=x[j]-0.5
          elif len(str(y[j]))==3:
            posx=x[j]-0.65
          elif len(str(y[j]))==4:  
            posx=x[j]-0.80
          ax.annotate(y[j],(x[j],y[j]), xytext=(posx,y[j]))
      
      ax.plot(totales.index,totales,c=cmap(i), label=t, marker="o", markersize=3)

    #Dibujar  
    plt.xticks(rotation=90, fontsize=12)
    plt.ylabel('Cantidad', fontdict = {'fontsize':20, 'color':'tab:blue'})  
    plt.xlabel('Año', fontdict = {'fontsize':20, 'color':'tab:blue'})
    plt.title( 'Cantidad películas x año de lanzamiento', loc = "center", fontdict = {'Weight':'bold', 'fontsize':20, 'color':'tab:blue'})
    plt.legend(loc='upper left', fancybox=True, framealpha=1, shadow=True, borderpad=1)
    plt.show()  

  except:
    print(f'Error en grafCantTipos: {s_error}')
    
````

````
grafCantTipos(df_Netflix_Filt,1925, 1945, True)
grafCantTipos(df_Netflix_Filt,1945, 1965, True)
grafCantTipos(df_Netflix_Filt,1965, 1985, True)
grafCantTipos(df_Netflix_Filt,1985, 2005, True)
grafCantTipos(df_Netflix_Filt,2005, 2022, True)
````

*Se analizará cantidad de peliculas y series subidas a Netflix por año*

````
def TotalTipoAnioSubida(df,anio_desde,anio_hasta):
  """
  Desc: Función que plotea las Cantidades por Tipo en el tiempo (de subida a Netflix)
  Args:
    df (dataframe): Dataframe depurado
    anio_desde: Año inicio de la escala en eje x
    anio_hasta: Año fin de la escala en eje x
  Returns:
    Plot por pantalla  
  """    
  import matplotlib.pyplot as plt

  try:

    fig, ax = plt.subplots(1, 1, figsize=(12, 6))
    color = ["Green", "Blue"]

    s_error=''
    for i, mtv in enumerate(df['type'].value_counts().index):
        mtv_rel = df[df['type']==mtv]['year_added'].value_counts().sort_index()
        ax.plot(mtv_rel.index, mtv_rel, color=color[i], label=mtv)
        ax.fill_between(mtv_rel.index, 0, mtv_rel, color=color[i], alpha=0.9)
        
    ax.yaxis.tick_right()
        
    ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha =0.7)

    s_error=''
    ax.legend(loc='center left',fontsize=15)
    for s in ['top', 'right','bottom','left']:
        ax.spines[s].set_visible(False)

    ax.grid(False)

    ax.set_xlim(anio_desde,anio_hasta)
    plt.xticks(np.arange(anio_desde,anio_hasta+1, 1))

    fig.text(0.15, 0.80, "Movies & TV Shows subidos a Netflix por Año ("+ str(anio_desde) +" - " + str(anio_hasta) +")", fontsize=15)
    ax.tick_params(axis=u'both', which=u'both',length=0)

    plt.show()
  except:
    print(f'Error en TotalTipoAnioSubida: {s_error}')   

TotalTipoAnioSubida(df_Netflix_Filt, 2010, 2021)  
````

Cantidad Peliculas: Por pais y año lanzamiento (Top)

````
def TotalPaisAnio(df,cantidad,anio_rel_desde,anio_rel_hasta):
  """
  Desc: Función que plotea las Cantidades Totales por Pais y Anio Lanzamiento
  Args:
    df (dataframe): Dataframe depurado
    cantidad: Top de paises solicitado
    anio_rel_desde: Año de Lanzamiento para inicio de la escala en eje x
    anio_rel_hasta: Año de Lanzamiento para fin de la escala en eje x
  Returns:
    Plot por pantalla  
  """  
  import matplotlib.pyplot as plt
  fig, ax = plt.subplots(1, 1, figsize=(12, 6))

  for i, pais in enumerate(df['country'].value_counts()[:cantidad].index):
      pais_rel = df[df['country']==pais]['release_year'].value_counts().sort_index()
      ax.plot(pais_rel.index, pais_rel, label=pais)
      
  ax.yaxis.tick_right()
  ax.axhline(y = 0, color = 'black', linewidth = 1.3, alpha =0.7)

  ax.legend(loc='center left',fontsize=15)
  for s in ['top', 'right','bottom','left']:
      ax.spines[s].set_visible(False)

  ax.grid(False)

  ax.set_xlim(anio_rel_desde,anio_rel_hasta)
  plt.xticks(np.arange(2000, 2022, 1))

  fig.text(0.15, 0.80, "Top "+str(cantidad)+" paises: Por cantidad y año lanzamiento ("+ str(anio_rel_desde) + " - " + str(anio_rel_hasta) + ")", fontsize=15)
  ax.tick_params(axis=u'both', which=u'both',length=0)

  plt.show()

TotalPaisAnio(df_Netflix_Filt, 5, 2000, 2021)
````

````
def lista_valores(df, column):
  """
  Desc: Función que dado un data frame y una columna que tiene un string separado por coma, retorna una lista
  Args:
    df: data frame
    columna: columna del data frame que contienen un string con valores separados por coma

  Returns:
    lista con valores  
  """  
  lista_valores=[]
  lista=[]
  #try:

  for i in df.index:
    #separo el string en valores de una lista
    lista = df[column][i].split(',')
    for lista_valor in lista:
      #Tomo cada valor de la lista y si no esta en la lista de valores, lo agrego
      if lista_valor.strip() not in lista_valores :
        lista_valores.append(lista_valor.strip())
  #except: 
  #  print('Error al devolver los diferentes valores de un campo lista de un data frame.')
  return lista_valores

````

````
#cargo la lista de los diferentes generos llamando a la funcion lista_valores()
generos=[]
df_peliculas=df_Netflix[df_Netflix['type']=='Movie']
generos_p = lista_valores(df_peliculas, 'listed_in')
generos_p.sort()

df_series=df_Netflix[df_Netflix['type']=='TV Show']
generos_s = lista_valores(df_series, 'listed_in')
generos_s.sort()
````

````
def contarGenero(datos, genero, anio, type):
  """
  Desc: Función que dado el data frame de peliculas, un genero y un año, cuenta la cantidad de peliculas para ese genero y en ese año
  Args:
    datos: data frame de peliculas
    genero: un genero
    anio: un año
    type: un tipo de pelicula (Movie o TV Shows)
  Returns:
    la cantidad de peliculas que son del tipo indicado, pertenecen al genero y se lanzaron en el año indicado
  """  
  try:
    df_peliculas=datos[datos['type']==type] 
    df_peliculas=df_peliculas.query('@genero in listed_in')
    cant=len(df_peliculas[(df_peliculas['release_year']==anio)])
  except:
    print('Error al contar la cantidad de peliculas del tipo, genero y año indicado.')
  return cant
````

````
def dict_genero(lista_generos, lista_anios, df, tipo):
  """
  Desc: Función que dado el data frame de peliculas, una lista de generos, una lista de años y un tipo, 
  genera un diccionario con clave el genero como clave que contiene diccionarios donde la clave es el año y el valor la cantidad de peliculas en ese año
  Args:
    lista_generos: lista de generos
    lista_anios: lista de años
    df: data frame de peliculas
    type: un tipo de pelicula (Movie o TV Shows)
  Returns:
    un diccionario 
  """  

  try:
    dict_genero={}
    for gen in lista_generos:
        sub_dic={}
        for an in lista_anios:
            cant= contarGenero(df, gen, an, tipo)
            sub_dic[an] = cant
        dict_genero[gen] = sub_dic
  except:
    print('Error al construir el diccionario de generos')  
  return dict_genero
````

````
def grafica_por_genero(desde_anio, hasta_anio, generosAMostrar, gen_peliculas, df, tipo, colores):
  """
  Desc: Función que realiza 4 plots con 4 generos que se pasan por parametro, entre los años indicados
  Args:
    desde_anio: año a partir del cual se mostraran los plots
    hasta_anio: año hasta el cual se mostraran los plots
    generosAMostrar: lista con los generos que se quieren plotear
    gen_peliculas: generos de peliculas
    df (dataframe): Dataframe de peliculas
    tipo: un tipo de pelicula (Movie o TV Shows)
    colores: lista de colores

  Returns:
    Plot por pantalla  
  """  
  import matplotlib.pyplot as plt  
  try:
  
    if len(generosAMostrar) != 4:
      print('Debe ingresar 4 generos')
    else:
      mask2 = (df['release_year'] >= desde_anio) & (df['release_year'] <= hasta_anio)
      df = df.loc[mask2]

      anios=[]
      anios = list(df['release_year'].drop_duplicates())
      anios.sort()

      dic_genero = dict_genero(gen_peliculas, anios, df, tipo)

      fig = plt.figure(figsize=(20,12))
      ax = fig.subplots(nrows=2, ncols=2)

      count_nrow = 0
      count_ncols = 0
      i = 0

      for genero in generosAMostrar:
        ax[count_nrow, count_ncols].plot(anios, list(dic_genero[genero].values()))
        ax[count_nrow, count_ncols].set_title(genero)
        ax[count_nrow, count_ncols].fill_between(anios, 0, list(dic_genero[genero].values()), color=colores[i])
        i += 1
        if count_nrow == 0 and count_ncols == 0:
          count_ncols = 1
        elif count_nrow == 0 and count_ncols == 1:
          count_nrow = 1
          count_ncols = 0
        elif count_nrow == 1 and count_ncols == 0:
          count_nrow = 1
          count_ncols = 1

      plt.show()

  except:
    print('Error al dibujar la grafica de los generos')  
````
